# Premier League Player Clustering

I am using Premier League player data from the 2024/25 season to run some clustering analysis. I will run the analysis:

a) for all outfield players together 

b) for players in each position separately

By running the analysis for all outfield players, clusters are likely to largely relate to playing position. By running the analysis for each position separately, the results have potential to be far more interesting as they could highlight specific attributes or tactical nuances.

The data is taken after 25 matches of the season and all statistics are on a per 90 minutes basis.

Data sourced from the following site: https://fbref.com/en/comps/9/Premier-League-Stats

```python
# Import Libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
import warnings
```

```python
# Load Data
standard = pd.read_csv('Standard.csv', encoding='latin1')
shooting = pd.read_csv('Shooting.csv', encoding='latin1')
possession = pd.read_csv('Possesion.csv', encoding='latin1')
passing_types = pd.read_csv('PassingTypes.csv', encoding='latin1')
passing = pd.read_csv('Passing.csv', encoding='latin1')
misc = pd.read_csv('Misc.csv', encoding='latin1')
goal_creation = pd.read_csv('GoalShotCreation.csv', encoding='latin1')
defensive = pd.read_csv('Defensive.csv', encoding='latin1')
```

## Merge Data

- Created a main dataframe from all of the individual ones
- Started using the "standard" dataset, removed goalkeepers and players who hadn't played more than 450 minutes
- Updated position so that each player has only one position - applied arbitrary rule
- Specified which columns from each dataframe should be included in main dataframe - necessary to avoid duplication and removes some unnecessary noise
- Joined each dataframe to the main dataframe using a for loop

```python
# create main df
main_df = standard[(standard['90s']>=5) & (standard['Pos'] != 'GK')].reset_index()

# update positions
def update_position(row):
    if 'FW' in row['Pos']:
        return 'FW'
    elif 'MF' in row['Pos']:
        return 'MF'
    else:
        return 'DF'

main_df['Pos'] = main_df.apply(update_position, axis=1)

# filter out columns
standard_cols = [
    'Player',
    'Squad',
    'Pos',
    'Gls', #goals
    'G-PK', #non-penalty goals
    'PK', #penalty goals
    'Ast', #assists
    'CrdY', #yellows
    'CrdR', #reds
    'xG', #expected goals
    'npxG', #non-penalty expected goals
    'xAG', #expected assists
    'PrgC', #progressive carries
    'PrgP', #progressive passes
    'PrgR' #progressive passes received
]

main_df = main_df[standard_cols]

# select columns for each stat area
shooting_cols = [
    'Player',
    'Squad',
    'Sh', #shots
    'SoT', #shots on target
    'Dist', #avg shot distance
    'FK' #shots from free kicks
    
]

possession_cols = [
    'Player',
    'Squad',
    'Touches', 
    'Def Pen', #touches in own box
    'Def 3rd', #touches in defensive third
    'Mid 3rd', #touches in middle third
    'Att 3rd', #touches in attacking third
    'Att Pen', #touches in opposition box
    'Live', #live-ball touches (i.e. not set piece)
    'Att', #attempted take ons
    'Succ', #successful take ons
    'Succ%', #successful take on %
    'Tkld', #times tackled during take on
    'Tkld%', #% of times tackled during take on
    'Carries', #times controlled ball with their feet
    'TotDist', #distance moved while controlling ball
    'PrgDist', #as above but in direction of opponent's goal
    #'1/3', #carries into final third
    'CPA', #carries into penalty box
    'Mis', #miscontrols
    'Dis', #times dispossessed
    'Rec' #passes received
]

pt_cols = [
    'Player',
    'Squad',
    'Live', #live-ball passes
    'Dead', #dead-ball passes
    'FK', #free kicks
    'TB', #through balls
    'Sw', #switches
    'Crs', #crosses
    'TI', #throw-ins
    'CK', #corners
    'Off', #offside passes
    'Blocks' #blocked passes
]

passing_cols = [
    'Player',
    'Squad',
    'Cmp', #completed passes
    'Att', #attempted passes
    'Cmp%', #completion %
    'TotDist', #distance completed
    'PrgDist', #distance completetd forwards
    'Cmp.1', #short passes complete (5-15 yd)
    'Att.1', #short passes attempted
    'Cmp%.1', #short pass completion %
    'Cmp.2', #mid distance (15-30 yd)
    'Att.2',
    'Cmp%.2',
    'Cmp.3', #long distance (>30 yd)
    'Att.3',
    'Cmp%.3',
    'KP', #key passes
    'PPA', #passes into box
    'CrsPA', #crosses into box
    #'PrgP' #forward passes
]

misc_cols = [
    'Player',
    'Squad',
    'Fls', #fouls committed
    'Fld', #times fouled
    'Off', #offside
    'OG', #own goals
    'Recov', #ball recoveries
    'Won', #aerial duals won
    'Lost', #aerial duals lost
    'Won%' #aerial dual win %
]

gc_cols = [
    'Player',
    'Squad',
    'SCA', #shot creating actions
    'PassLive', #Live-ball SCA
    'PassDead', #Dead-ball SCA
    'TO', #SCA with successful take on
    'Sh', #SCA with shot
    'Fld', #SCA with foul drawn
    'Def', #SCA with defensive action
    'GCA', #goal creating action
    'PassLive.1', #Live-ball GCA
    'PassDead.1', #Dead-ball GCA
    'TO.1', #GCA with successful take on
    'Sh.1', #GCA with shot
    'Fld.1', #GCA with foul drawn
    'Def.1', #GCA with defensive action
]

def_cols = [
    'Player',
    'Squad',
    'Tkl', #players tackled
    'TklW', #tackles won
    'Def 3rd', #tackles in defensive third
    'Mid 3rd', #tackles in middle third
    'Att 3rd', # tackles in attacking third
    'Tkl.1', #dribblers tackled
    'Att', #dribblers challenged
    'Tkl%', #% dribblers tackled
    'Lost', #unsuccessful attempts to stop a dribbler
    'Blocks', #no of blocks
    'Sh', #shots blocked
    'Pass', #passes blocked
    'Int', #interceptions
    'Clr', #clearances
    'Err' #error leading to a shot
]

# update fields in each df
shooting = shooting[shooting_cols]
possession = possession[possession_cols]
passing = passing[passing_cols]
passing_types = passing_types[pt_cols]
misc = misc[misc_cols]
goal_creation = goal_creation[gc_cols]
defensive = defensive[def_cols]

# join each df to main df
data = [
    shooting,
    possession,
    passing,
    passing_types,
    misc,
    goal_creation,
    defensive,
]

for df in data:
    main_df = pd.merge(
        main_df,
        df,
        on = ['Player','Squad'],
        how = 'left'
    )

main_df.head()
```

## Principal Component Analysis

There are 99 columns in the dataset. By applying PCA we create a dataset with new features (principal components) that are linear combinations of the original features. This will result in increased computational efficiency and reduce the amount of noise in the datatset. The below "elbow chart" shows that the first three principal components account for over 50% of the variation in the datatset.

```python
explained_variances = []
for n_components in range(0, 30):
    pca = PCA(n_components=n_components)
    pca.fit(scaled_features_df)
    explained_variances.append(np.sum(pca.explained_variance_ratio_))

# Plot the Elbow Chart
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variances) + 1), explained_variances, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Elbow Method for Optimal Number of Principal Components')
plt.grid(True)
plt.show()
```

The below code creates a new dataset - I have specified that I want the new dataset to explain 80% of the variation in the original sample. The result has 17 principal components rather than the 99 original features.

```python
pca = PCA(0.8)
features_pca = pca.fit_transform(scaled_features_df)
features_pca.shape
```

The first principal component explains over 30% of the total variance, the second 14% and the third 8%.

```python
pca.explained_variance_ratio_
```

The below code generates the "loading scores" (i.e. coefficients) of each of the original fields for each principal component. The top 5 loading scores for each of the first 5 principal components are listed below.

```python
# Get the loading scores (principal component coefficients)
loading_scores = pca.components_

# create index 
pc_index = [f"PC{i+1}" for i in range(features_pca.shape[1])]

# Create a DataFrame to display the loading scores
loading_df = pd.DataFrame(loading_scores, columns=main_df_features.columns, index=pc_index)

# Print the loading scores
print(loading_df)

```

```python
for i in range(5):
    loading = loading_df.iloc[i]
    sorted_loading = loading.abs().sort_values(ascending=False)
    print(f"\nFeatures contributing to PC{i + 1}:")
    for feature, score in sorted_loading.head().items():
        print(f"  - {feature}: {score:.3f}")
```

## K-Means Clustering

### All positions

```python
def plot_elbow(df):
    SSE = []
    for cluster in range(1,10):
        kmeans = KMeans(n_clusters = cluster, init = 'k-means++', random_state=61)
        kmeans.fit(df)
        SSE.append(kmeans.inertia_)

    frame = pd.DataFrame({'Cluster': range(1,10), 'SSE': SSE})
    plt.figure(figsize=(12,6))
    plt.plot(frame['Cluster'], frame['SSE'], marker = 'o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')

    plt.show()
```

The elbow plot can be used for K-Means clustering to identify the optimal number of clusters to achieve the best degree of separation. Using the original datatset, there is a clear point of inflexion at n=2, however the result is unlikely to be very interesting - it will probably just separate attacking players from defensive players.

```python
plot_elbow(scaled_features_df)
```

There is also a point of inflexion at k=4. I fitted four clusters and computed the silhouette score, which is a measure of the degree of separartion achieved. The silhouette score achieved was around 16%, which indicates a poor level of separation in the data.

```python
kmeans = KMeans(n_clusters = 4, init='k-means++', random_state=61)
kmeans.fit(scaled_features_df)
print(silhouette_score(scaled_features_df, kmeans.labels_, metric = 'euclidean'))
```

Using the dataset from PCA and k=4, I achieved a silhouette score of 0.21 which is still quite low but shows that we have achieved more separation by removing noise from the dataset.

```python
plot_elbow(features_pca)
```

```python
kmeans = KMeans(n_clusters = 4, init='k-means++', random_state=61)
kmeans.fit(features_pca)
print(silhouette_score(features_pca, kmeans.labels_, metric = 'euclidean'))
```

Using four clusters, I fitted my clusters and then joined the cluster back to the scaled dataframe so that we can identify which players make up each cluster.

```python
#predict cluster using PCA
cluster = kmeans.fit_predict(features_pca)
print(cluster)
```

## Summary of Findings

### All Players

When analysing all players together, players could be separated into the following four clusters:

- Cluster 0 - Central Strikers
- Cluster 1 - Central Defenders
- Cluster 2 - Central Midfielders and Full Backs
- Cluster 3 - Creative Players

Some of the findings in practical terms:
- Players are separated largely along positional lines
- Central defenders have more touches in their own penalty box and defensive third and play a large number of longer and forwards passes
- Forwards can be separated out into cluster 0 or cluster 3 depending on whether they create chances or are on the end of them
- Central midfielders and full backs tend to be put into the same cluster. For this cluster, there are no features where the mean for the cluster is more than one standard deviation from the mean for all players. This may reflect that these positions balance attacking and defensive responsibilities.

### Forwards

- Cluster 0 - Misclassified Forwards and Ineffective Forwards
- Cluster 1 - Wide Players
- Cluster 2 - Central Strikers

Cluster 0 does not have any attacking metrics that stand out compared to all other players. Cluster 1 and Cluster 2 have different attacking metrics - players in cluster 1 have more take-ons and carry the ball into the penalty area more; players in cluster 2 have more shots on target and expected goals.

### Midfielders

- Cluster 0 - Deep Lying Midfielders
- Cluster 1 - Creative Midfielders

Cluster 0 players make more tackles in the middle third. Cluster 1 players are ahead on a number of attacking metrics. There are far more midfielders in Cluster 0 than Cluster 1 - this suggests that the "number 10" may be a dying breed as more teams are playing 4-3-3 formations with their wide forwards as the main creative outlets.

### Defenders

- Cluster 0 - Central Defenders (Top Half Clubs)
- Cluster 1 - Central Defenders (Bottom Half Clubs)
- Cluster 2 - Central Midfielders and Full Backs

The top half/ bottom half distinction is a bit of a generalisation, but on the whole the better sides have their centre backs complete more progressive passes, while the bottom half teams have their centre backs make more clearances and block more shots.
